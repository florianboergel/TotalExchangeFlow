```python
import xarray as xr
from pyTEF.tef_core import constructorTEF
```

```python
ds = xr.open_mfdataset("/fast/boergel/baltic_sea_950_1800_amo_ergom/model_simulations/MOM_950_1800_FACTOR_3/1000/ocean_day3d.nc",
                     use_cftime=True).load()
ds_vel = xr.open_mfdataset("/fast/boergel/baltic_sea_950_1800_amo_ergom/model_simulations/MOM_950_1800_FACTOR_3/1000/ocean_trps.nc",
                     use_cftime=True).load()
```

```python
ds_vel_transect = ds_vel.tx_trans.sel(xu_ocean=13, method="nearest")
```

```python
ds_transect = ds.sel(xt_ocean = 13, method = "nearest")
```

```python
ds_transect_salt = ds_transect.salt.to_dataset(name="salt")
```

```python
ds_transect_salt["tx_trans"]= ds_vel_transect/1000
```

```python
data_description = {
    "lon" : None,
    "lat" : "yt_ocean",
    "depth" : "st_ocean",
    "time" : "time"
}
```

```python
tef = constructorTEF(inputFile=ds_transect_salt,
                     data_description=data_description)
```

```python
from pyTEF.calc import sort_1dim, sort_2dim
```

```python
tef.tracer = tef.ds.salt
tef.transport = tef.ds.tx_trans
```

```python
sort_1dim(tef)
```

```python
tef.tracer = [tef.ds.salt, tef.ds.salt]
```

```python
sort_2dim(tef)
```
